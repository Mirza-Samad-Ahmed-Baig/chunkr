```python
import asyncio
import tracemalloc
import nest_asyncio
from chunkr_ai import Chunkr

# Enable tracemalloc for debugging memory issues
tracemalloc.start()

# Fix for Jupyter Notebook's running event loop
nest_asyncio.apply()

url = "https://chunkr-web.s3.us-east-1.amazonaws.com/landing_page/input/science.pdf"

async def process_document():
    # Initialize client with API key
    chunkr = Chunkr(api_key="Your_api_key")

    try:
        # Upload a file and wait for processing
        task = await chunkr.upload(url)
        print(f"Task ID: {task.task_id}")

        # Create task without waiting (alternative method)
        task = await chunkr.create_task(url)


        # Poll the task until completion
        await task.poll()
        task.markdown(output_file="output.md")
        task.html(output_file="output.html")
        print("Document Chunks:", task.output.chunks)

    except Exception as e:
        print(f"Error: {e}")

    finally:
        # Ensure proper cleanup of resources
        await chunkr.close()

# Run the async function in Jupyter
await process_document()
```
